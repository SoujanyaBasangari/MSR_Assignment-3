```python
import requests
from bs4 import BeautifulSoup
import re
session1 = requests.Session()
url = "https://paperswithcode.com/sota"
task_links=[]
req1 = session1.get(url)
bs_2 = BeautifulSoup(req1.text, 'html.parser')
#     # bs.find_all('a')
# # bs.prettify()
#     sota = bs_2.findAll('table',{"class":"table-striped table-responsive"})[0]
# # print(table)
#     sota_links=sota.findAll('a')
sota_links=bs_2.findAll('a', href = re.compile(r'/task/.*'))
    
for anchor in sota_links:
    href = anchor['href'] 
    href = 'https://paperswithcode.com' + href
    task_links.append(href)
print(task_links)
# len(task_links)
# task_links[]
```

```python
session = requests.Session()
# html = ['https://paperswithcode.com/task/semantic-segmentation', 'https://paperswithcode.com/task/image-classification', 'https://paperswithcode.com/task/object-detection', 'https://paperswithcode.com/task/image-generation', 'https://paperswithcode.com/task/denoising', 'https://paperswithcode.com/task/machine-translation', 'https://paperswithcode.com/task/language-modelling', 'https://paperswithcode.com/task/question-answering', 'https://paperswithcode.com/task/sentiment-analysis', 'https://paperswithcode.com/task/text-generation', 'https://paperswithcode.com/task/medical-image-segmentation', 'https://paperswithcode.com/task/drug-discovery', 'https://paperswithcode.com/task/lesion-segmentation', 'https://paperswithcode.com/task/brain-tumor-segmentation', 'https://paperswithcode.com/task/covid-19-detection', 'https://paperswithcode.com/task/regression', 'https://paperswithcode.com/task/representation-learning', 'https://paperswithcode.com/task/transfer-learning', 'https://paperswithcode.com/task/word-embeddings', 'https://paperswithcode.com/task/data-augmentation', 'https://paperswithcode.com/task/recommendation-systems', 'https://paperswithcode.com/task/fairness', 'https://paperswithcode.com/task/continual-learning', 'https://paperswithcode.com/task/topic-models', 'https://paperswithcode.com/task/causal-inference', 'https://paperswithcode.com/task/semantic-segmentation', 'https://paperswithcode.com/task/link-prediction', 'https://paperswithcode.com/task/node-classification', 'https://paperswithcode.com/task/graph-embedding', 'https://paperswithcode.com/task/graph-classification']
paper=[]
for x in task_links[0:15]:
    req = session.get(x)
    bs = BeautifulSoup(req.text, 'html.parser')
    # bs.find_all('a')
# bs.prettify()
    table = bs.findAll('table',{"class":"table-striped table-responsive"})[0]
# print(table)
    links=table.findAll('a')
    links1=table.findAll('a', href = re.compile(r'/paper/.*'))
    
    for anchor in links1:
        href = anchor['href'] 
        href = 'https://paperswithcode.com' + href
        paper.append(href)
print(paper)
```

```python
final_list=[]
title=[]
for idp,x in enumerate(paper):
    #create new list
    list1=[]
    
    request = session.get(x)
    bs_1 = BeautifulSoup(request.text, 'html.parser')
    title= bs_1.find('div', class_='paper-title').h1
    #print(title.text)
    title.append(title.text)
    list1.append(title.text)
#     content = str(title)
#     print(title)
    task = bs_1.findAll('a', href = re.compile(r'/task/*'))
    t=[]
    for z in task:
        #print(z.text,idp)
        
        t.append(z.text)
    t = [x.replace('\n', '') for x in t]
    t = list(set(t))
    list1.append(t)
        #append1
    dataset = bs_1.findAll('a', href = re.compile(r'/dataset/*'))
    t2=[]
    for m in dataset:
        
        t2.append(m.text)
        #print(m.text,idp)
#         t2 = t2.remove('c')
    del t2[0]
    t2 = [x.replace('\n\n                                      Add a new dataset here\n                                  ', '') for x in t2]
    t2 = [x.replace('\n', '') for x in t2]
    t2 = [x.replace('  ', '') for x in t2]
    t2 = list(set(t2))
    list1.append(t2)
    git_code = bs_1.findAll('a', href = re.compile(r'https://github.com/.*'))
#     print(git_code)
    t3=[]
    for y in git_code:
        
        a_href=y['href']
        t3.append(a_href)
    t3 = [x.replace('\n', '') for x in t3]
    t3 = list(set(t3))
    list1.append(t3)
        #print(a_href,idp)
    final_list.append(list1)
        
```

```python
import pandas as pd
df = pd.DataFrame(final_list, columns=['Title','Domain','Datasets','GIT link'])
```

```python
table = df.replace(r'\n',  '', regex=True)
table = table.replace(r'  ',  '', regex=True)
# print(table)
```

```python
table=table.drop_duplicates(subset='Title', keep="last")
```
